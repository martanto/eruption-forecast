```python
import pandas as pd
from datetime import datetime, timedelta
```

```python
eruptions = [
    "2025-03-20",
    "2025-04-22",
    "2025-05-18",
    "2025-06-17",
    "2025-07-07",
    "2025-08-01",
    "2025-08-17",
]
```

```python
start_date = "2025-01-01"
end_date = "2025-12-31"
window_size = 2
window_overlap = 0.75
sampling_rate = 100.0
day_to_forecast = 2
```

```python
number_of_sample_points_per_day = sampling_rate * 3600 * 24
```

```python
total_sample_point = number_of_sample_points_per_day * window_size
```

```python
non_overlap_sample_points = (
    total_sample_point
    if window_overlap == 1.0
    else (1 - window_overlap) * total_sample_point
)
```

```python
non_overlap_in_seconds = timedelta(seconds=non_overlap_sample_points / sampling_rate)
non_overlap_in_seconds
```

```python
dates = pd.date_range(start=start_date, end=end_date, freq=non_overlap_in_seconds)
```

```python
df = pd.DataFrame(index=dates)
df["is_erupted"] = False
```

```python
for eruption in eruptions:
    end_eruption = datetime.strptime(eruption, "%Y-%m-%d")
    start_eruption = end_eruption - timedelta(days=day_to_forecast)
    for index, row in df.loc[start_eruption:end_eruption].iterrows():
        df.loc[index, "is_erupted"] = True
```

```python
df["time"] = df.index
df.reset_index(drop=True, inplace=True)
df["id"] = df.index
df = df[["id", "time", "is_erupted"]]
```
